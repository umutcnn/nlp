```python
#import libraries
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
#model ve tokenizer yükle
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
#input text (metni) tanimla
text = "Transformers can be used gor natural language prosessing."
```

```python
# metni tokenlere çevir
inputs = tokenizer(text, return_tensors="pt") # çıktı pytorch tensoru olarak return edilir
```

```python
with torch.no_grad():  # gradyan hesaplamayı kapat, daha hızlı ve hafif olur
    outputs = model(**inputs)
```

```python
#modelin çıkışından son gizli durumu alalım 
last_hidden_state = outputs.last_hidden_state #tun token çıktılarını almak için
```

```python
#ilk tokenin embedding ini alalım ve print ettirelim
first_token_embedding = last_hidden_state[0,0,:].numpy()

print(f"metin temsii : {first_token_embedding}")
```
